```python
## 1.text documents loader
```

```python
from langchain_community.document_loaders import TextLoader
from dotenv import load_dotenv

load_dotenv()

# Document loader part only
loader = TextLoader('cricket.txt', encoding='utf-8')
docs = loader.load()

# Document analysis only
print("Document Type:", type(docs))
print("Number of Documents:", len(docs))
print("Content:", docs[0].page_content)
print("Metadata:", docs[0].metadata)
```

```python
## 2 pdf loader
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader('git.pdf')

docs = loader.load()

print(len(docs))

print(docs[0].page_content)
print(docs[1].metadata)
```

```python
## 4 directory loader
```

```python
from langchain_community.document_loaders import DirectoryLoader

# Load ALL .txt files from a directory
loader = DirectoryLoader(
    path="./my_documents/",  # Directory path
    glob="*.txt",           # Pattern to match files
    show_progress=True      # Show loading progress
)

docs = loader.load()
print(f"Loaded {len(docs)} documents from the directory")
```
